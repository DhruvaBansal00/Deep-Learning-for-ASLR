```
import glob

def writeList(arkLoc: str, listFile: str, listLabel: str):
    files = glob.glob(arkLoc)
    print(files)
    listF = open(listFile, "w")
    allFiles = "\n".join(files)
    listF.write(allFiles)
    fileNames = [" ".join(i.split("/")[-1].split(".")[1].split("_"))+"\n" for i in files]
    listL = open(listLabel, "w")
    listL.writelines(fileNames)


writeList("/content/drive/My Drive/Deep-Learning-for-ASLR/data/ark/*", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/train.data", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/train.en")
writeList("/content/drive/My Drive/Deep-Learning-for-ASLR/data/ark/*", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/dev.data", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/dev.en")
writeList("/content/drive/My Drive/Deep-Learning-for-ASLR/data/ark/*", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/test.data", "/content/drive/My Drive/Deep-Learning-for-ASLR/data/lists/test.en")
```
